```python
# importing dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
iris =load_iris()

#converting into a pd dataframe
iris_df = pd.DataFrame(data = iris.data, columns = iris.feature_names)

#make a seperate target clmn representing target values
iris_df['target'] = iris.target

print(iris_df.head())
```

```python
iris_df.shape
```

```python
iris_df.info()
```

# **Data Visualization**

```python
sns.set_theme()
```

```python
sns.scatterplot(data = iris_df, x = 'sepal length (cm)', y = 'petal width (cm)', hue = 'target')
```

```python
correlation = iris_df.corr()
```

```python
# figure size
plt.figure(figsize=(5, 6))

#  heatmap of correlation matrix
heatmap = sns.heatmap(correlation, cmap='coolwarm', annot=True, fmt='.2f', linewidths=.9, square=True)

plt.title('Correlation Heatmap', fontsize=12)

plt.show()
```

```python
X = iris_df.drop(columns=['sepal width (cm)','petal length (cm)'], axis =1)

Y = iris_df['target'].to_numpy()

# Y represents the target columns
```

# **Model Training**

```python

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
print(Y.shape, y_train.shape, y_test.shape)
```

```python
model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, prediction)

accuracy = accuracy * 100
```

```python
print(f"Accuracy: {accuracy}%")
```

```python
plt.figure(figsize=(5, 7))
plt.bar(['Accuracy'], [accuracy], color='skyblue')
plt.ylim(0, 100)  # Set the y-axis limit to 0-100 for percentage
plt.xlabel('Metrics')
plt.ylabel('Percentage (%)')
#plt.title('Model Accuracy')

# Display the accuracy value on top of the bar
plt.text('Accuracy', accuracy + 2, f'{accuracy:.2f}%', fontsize=12)

plt.show()
```
